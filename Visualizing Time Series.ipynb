```python
import pandas as pd  
path   = 'london_weather_MODIFIED - london_weather_MODIFIED (1).csv'
df = pd.read_csv(path)
df.info ()
```

```python
pd.to_datetime(df['date'])
df.info()
```

```python
df.head()
```

```python
df['date']= pd.to_datetime(df['date'] , format='%Y%m%d')
df['date']
```

```python
# Set the date column as the index
df= df.set_index('date')
df
```

```python
df.index
```

```python
df=df.loc['2000':, ['max_temp','mean_temp','precipitation','min_temp','snow_depth']] 
df
```

```python
df['max_temp'] =df['max_temp'].interpolate()
df['mean_temp']= df['mean_temp'].interpolate()
df['min_temp']=df['min_temp'].interpolate()
      
```

```python
df['precipitation']= df['precipitation'].fillna(method='ffill')
df['snow_depth']= df['snow_depth'].fillna(0)
```

```python
df.isna().sum() 
```

```python
temp_18_20 = df.loc['2018':, ['max_temp', 'mean_temp','min_temp']]
temp_18_20
```

```python
df.index
```

```python
from matplotlib import dates as mdates
ax = temp_18_20.plot(figsize=(12,3))
ax.set(title = 'Daliy Temperature', ylabel= 'Degrees')
# customize major ticks
maj_loc = mdates.YearLocator(1)
maj_fmt = mdates.DateFormatter('%Y')
ax.xaxis.set_major_locator(maj_loc)
ax.xaxis.set_major_formatter(maj_fmt) 
# customize minor ticks
minor_loc =  mdates.MonthLocator(interval=3)
minor_fmt = mdates.DateFormatter("%b")
ax.xaxis.set_minor_locator(minor_loc)
ax.xaxis.set_minor_formatter(minor_fmt)
fig= ax.get_figure()
fig.autofmt_xdate(which='major',rotation=90)
```
