## Filter out only the required data:
   * Keep only data from the year 2000 or later.
      * Keep only the following features:
         * precipitation
         * mean_temp
         * min_temp
         * max_temp
         * snow_depth

```python
df=df.loc['2000'  ,['precipitation', 'mean_temp', 'min_temp', 'max_temp' , 'snow_depth'] ]
df
```

```python
date_low_temp = df['min_temp'].idxmax()
min_temp= df['min_temp'].max()
```

```python
print(f"The lowest temperature of {min_temp } degrees occured on {date_low_temp.strftime('%B %d, %Y')} which was a {date_low_temp.day_name()}.")
```
